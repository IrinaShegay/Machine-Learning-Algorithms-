```python
X_train_mean = x_train.copy()
X_test_mean = x_test.copy()

# Split the data into features (X) and target (y)
X = data.drop("target", axis=1)
y = data["target"]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)

# Create a copy of the dataframe to fill in empty cells with the mean value
X_train_mean = X_train.copy()
X_test_mean = X_test.copy()

# Fill empty cells with the mean value
for col in X_train_mean.columns:
    if X_train_mean[col].isnull().sum() > 0:
        mean = X_train_mean[col].mean()
        X_train_mean[col].fillna(mean, inplace=True)
        X_test_mean[col].fillna(mean, inplace=True)

# Create a logistic regression model
model = LogisticRegression()

# Train the model on the original data
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy without filling empty cells: ", accuracy)

# Train the model on the data with filled empty cells
model.fit(X_train_mean, y_train)

# Make predictions on the test set
y_pred_mean = model.predict(X_test_mean)

# Calculate the accuracy of the model
accuracy_mean = accuracy_score(y_test, y_pred_mean)
print("Accuracy with filling empty cells: ", accuracy_mean)
```
