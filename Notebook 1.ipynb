```
#Output the data of your group to the diagrams, while adding the name of the diagram: 
#subjects (bar chart),full name and grades (categorical chart)
import numpy as np
import matplotlib.pyplot as plt
 

barWidth = 0.20
fig = plt.subplots(figsize =(20, 8))
 

Python = [80, 60, 20, 70]
CPP = [70, 100, 90, 85]
Java = [75, 65, 70, 75]
FrontEnd = [95, 45, 70, 60]
 

Irina = np.arange(len(Python))
Damir = [x + barWidth for x in Irina]
Masha = [x + barWidth for x in Damir]
Alex = [x + barWidth for x in Masha]

plt.bar(Irina, Python, color ='r', width = barWidth,
        edgecolor ='grey', label ='Python')
plt.bar(Damir, CPP, color ='g', width = barWidth,
        edgecolor ='grey', label ='C++')
plt.bar(Masha, Java, color ='b', width = barWidth,
        edgecolor ='grey', label ='Java')
plt.bar(Alex, FrontEnd, color ='y', width = barWidth,
        edgecolor ='grey', label ='FrontEnd')
 

plt.xlabel('CS-2118', fontweight ='bold', fontsize = 25)
plt.ylabel('score', fontweight ='bold', fontsize = 25)
plt.xticks([r + barWidth for r in range(len(Python))],
        ['Irina', 'Damir', 'Masha', 'Alex'])
 
plt.legend()
plt.show()
```

```
#Plot the dependence of the weather data for the week by day using different lines.
import matplotlib.pyplot as plt
%matplotlib inline
Day = ['Mo', 'Tu', 'We', ' Th', 'Fr', 'Sa', 'Su']
maxtemp = [ 15, 18, 20, 22, 20, 17, 16]
mintemp = [5, 8, 10, 12, 10, 7, 6 ]
avgtemp = [10, 13, 15, 17, 15, 12, 11 ]

plt.ylabel('temperature', fontweight ='bold', fontsize = 20)
plt.xlabel('week', fontweight ='bold', fontsize = 20)
plt.plot( Day, maxtemp, '--', Day, mintemp, '-.', Day, avgtemp, ':')
```
